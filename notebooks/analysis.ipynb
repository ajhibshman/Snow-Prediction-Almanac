```python
#Notebook for analysis of weather data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
weather = "resources/Weather.csv"
weather_df = pd.read_csv(weather)
weather_df.head()
```

```python
#Aug precip to max

x_axis=weather_df['Aug_precip']
y_axis=weather_df['Max_SWE']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
plt.scatter(x_axis,y_axis,facecolor='r',edgecolor='b')
plt.plot(x_axis,regress_values,"g-")
plt.title('August Precipitation vs. Seasonal Max. SWE')
plt.xlabel('August Precipitation (in.)')
plt.ylabel('Season Max SWE (in.) ')
correlation = st.pearsonr(x_axis,y_axis)
print(f"The Correlation between August Precip and Total Season SWE is {round(correlation[0],2)}")
plt.savefig("../images/aug_precip_vs_max_swe.png")
plt.show
```
